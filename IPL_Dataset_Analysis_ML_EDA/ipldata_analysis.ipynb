```python
dm.team1.replace({'Rising Pune Supergiants' : 'Rising Pune Supergiant', 'Delhi Daredevils':'Delhi Capitals','Pune Warriors' : 'Rising Pune Supergiant'},inplace=True)
dm.team2.replace({'Rising Pune Supergiants' : 'Rising Pune Supergiant', 'Delhi Daredevils':'Delhi Capitals','Pune Warriors' : 'Rising Pune Supergiant'},inplace=True)
dm.toss_winner.replace({'Rising Pune Supergiants' : 'Rising Pune Supergiant', 'Delhi Daredevils':'Delhi Capitals','Pune Warriors' : 'Rising Pune Supergiant'},inplace=True)
dm.winner.replace({'Rising Pune Supergiants' : 'Rising Pune Supergiant', 'Delhi Daredevils':'Delhi Capitals','Pune Warriors' : 'Rising Pune Supergiant'},inplace=True)
dm.city.replace({'Bangalore' : 'Bengaluru'},inplace=True)
```

```python
print('\n Cities played at:', dm['city'].unique())
print('\n Teams participated:', dm['team1'].unique())
print('\n Teams participated:', dm['team2'].unique())
```

```python
dm['Season'] = pd.DatetimeIndex(dm['date']).year
dm.head()
```

```python
match_per_season = dm.groupby(['Season'])['id'].count().reset_index().rename(columns={'id':'matches'})
match_per_season
```

```python
plt.figure(figsize=(6,6))
sns.countplot(dm['Season'])
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('Season', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Total matches played in each season', fontsize = 12, fontweight = "bold")
```

```python
Cities = dm.groupby(['city'])['id'].count().reset_index().rename(columns={'id':'matches'})
Cities
```

```python
plt.figure(figsize=(17,6))
sns.countplot(dm['city'])
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('city', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Total matches played in each City', fontsize = 12, fontweight = "bold")
```

```python
season_data=dm[['id','Season']].merge(db, left_on = 'id', right_on = 'id', how = 'left').drop('id', axis = 1)
season_data.head()
```

```python
season=season_data.groupby(['Season'])['total_runs'].sum().reset_index()
p=season.set_index('Season')
ax = plt.axes()
ax.set(facecolor = "white")
sns.lineplot(data=p,palette="magma") 
plt.title('Total runs in each season',fontsize=16,fontweight="bold")
plt.show()
```

```python
winner = dm.groupby(['winner'])['id'].count().reset_index().rename(columns={'id':'matches'})
winner
```

```python
plt.figure(figsize=(15,6))
sns.countplot(dm['winner'])
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('winner', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Total matches win by each team', fontsize = 12, fontweight = "bold")
```

```python
runs_per_season=pd.concat([match_per_season,season.iloc[:,1]],axis=1)
runs_per_season['Runs scored per match']=runs_per_season['total_runs']/runs_per_season['matches']
runs_per_season.set_index('Season',inplace=True)
runs_per_season
```

```python
toss=dm['toss_winner'].value_counts()
ax = plt.axes()
ax.set(facecolor = "white")
sns.set(rc={'figure.figsize':(15,10)},style='darkgrid')
ax.set_title('No. of tosses won by each team',fontsize=15,fontweight="bold")
sns.barplot(y=toss.index, x=toss, orient='h',palette="icefire",saturation=1)
plt.xlabel('# of tosses won')
plt.ylabel('Teams')
plt.show()
```

```python
ax = plt.axes()
ax.set(facecolor = "white")
sns.countplot(x='Season', hue='toss_decision', data=dm,palette="colorblind",saturation=1)
plt.xticks(rotation=90,fontsize=10)
plt.yticks(fontsize=15)
plt.xlabel('\n Season',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.title('Toss decision across seasons',fontsize=12,fontweight="bold")
plt.show()
```
