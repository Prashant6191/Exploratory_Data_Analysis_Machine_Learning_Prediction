#year wise data of released tv shows and movies

```python
df['Date']=pd.to_datetime(df['Release_Date'])
```

```python
df.head(5)
```

```python
df['Date'].dt.year.value_counts()
```

#find in which year highest number of tv shows and movies were released.show with bar plot also

```python
plt.figure(figsize=(9,9))
df["Date"].dt.year.value_counts().plot(kind="bar")
plt.title("tv-shows and movies bar plot")
plt.ylabel("Count")
plt.xlabel("Year")
plt.show()
```

#what type of contents are present in the dataset.

```python
#what type of contents are present in the dataset.
plt.figure(figsize=(10,5))
plt.pie(df['Category'].value_counts().sort_values(),labels=df['Category'].value_counts().index,explode=[0.05,0],autopct='%1.2f%%',colors=['red','yellow'])
plt.show()
```

# Top 10 Countries with highest no of movies and TV Shows 

```python
plt.figure(figsize=(10,5))
plt.title("Top 10 Countries with highest no of movies and TV Shows ")
plt.xticks(rotation=60)
sns.countplot(x = df.Country, order = df['Country'].value_counts().index[0:10]);
```
