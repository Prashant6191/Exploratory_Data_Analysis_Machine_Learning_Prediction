#### Here we can see uniques value of 'RATING_TYPE' in which 'Excellent', 'Very Good', 'Good','Average', 'Poor', 'None', 'Not rated', are valid rating. except these all value are not valid we replace these value with 'Average' value.

```python
def clean_data(data):
  if (data=='Excellent') or (data=='Very Good') or (data=='Good') or (data=='Average') or (data=='Poor') or (data=='None') or (data=='Not rated'):
    return data
  else:
    data='Average'
    return data
```

```python
df['rating_text']=df['rating_text'].map(clean_data)
df['rating_text'].unique()
```

####now we rectify 'Rating_text' column. we can see in these uniques value all are valid.

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
# which City has Maximum and Minimum number of Restaurant?
x=df['city'].value_counts()
print(f'{x.index.tolist()[0]} City has Maximum number of {x[0]} and {x.index.tolist()[-1]} City has Minimum number of {x[-1]} Restaurant')
plt.figure(figsize=(10,5))
x.plot(kind='bar')
plt.xlabel("CITY")
plt.ylabel("NUmber of Restaurant")
plt.show()
```

```python
# which CUSINE TYPE have maximum number of restaurant in india?
x=df['cuisines'].mode()
print(f'{x.values[0]} CUSINE TYPE have maximum number of restaurant in india')
plt.figure(figsize=(20,25))
df['cuisines'].value_counts().plot(kind='barh')
plt.xlabel("CUSINE TYPE")
plt.ylabel("NUmber of Restaurant")
plt.show()
```

```python
#which Restaurant has highest rating and Rating type is Excellent in Agra City?
x=df['name'].loc[(df['rating_text']=='Excellent') | (df['aggregate_rating']==df['aggregate_rating'].max()) | (df['city']=='Agra')].values
print(f'In Agra City Restaurant {x[0]}, {x[1]}, {x[2]},\n {x[3]} and {x[4]} has highest rating and Rating type is Excellent.')
```

```python
# Which Restaurant In Ahmadabad city has price less than 2.0, RATING more than 3.6 and CUSINE CATEGORY is atleast North Indian & Chinese?
x=df[(df['city']=='Ahmedabad') & (df['price_range']<=2.0) & (df['aggregate_rating']>3.6)]
for i in x['name']:
  if 'North Indian' in x['cuisines'].loc[x['name']==i].values[0] and 'Chinese' in x['cuisines'].loc[x['name']==i].values[0]:
    print(f'{i}' ,end='\n ')
```

```python
# which Restaurant in the Region have more than 100 Votes and RATING_TYPE is Average or more than that with Price is equal or greater than 4.0.
x=df[(df['city']=='Ahmedabad') | (df['votes']>=100) & ((df['rating_text']=='Very Good') | (df['rating_text']=='Good') | (df['rating_text']=='Excellent') | (df['rating_text']=='Average')) & (df['price_range']>=4.0)]
y=x['name'].values
print(f"{y[0]}, {y[1]}, {y[2]} and {y[3]} are the Restaurant in the 'Ahmedabad' Region \n have more than 100 Votes and RATING_TYPE is Average or more than that with Price is equal or greater than 4.0.")
```

```python
#In the gandhinagar City which Region get highest and lowest Votes and Restaurant have Rating more than 4?
x=df[(df['city']=='gandhinagar') | (df['aggregate_rating']> 4)]
y=x[['name','locality']].loc[x['votes']==x['votes'].max()].values
z=x[['name','locality']].loc[x['votes']==x['votes'].min()].values
print(f"In the gandhinagar City {y[0][1]} Region {y[0][0]} Restaurant get highest Votes with more than 4 votes whareas {z[0][1]} Region {z[0][0]} Restaurant get lowest votes with more than 4 votes")
plt.figure(figsize=(25,10))
plt.bar(x['name'],x['votes'])
plt.xlabel("NAME")
plt.ylabel("VOTES")
plt.xticks(rotation=90)
plt.show()
```

```python
# which city restaurant get higest Votes and lowest votes with price is more than 3.0.
x=df[df['price_range']>=3.0]
y=x.groupby(['city'])['votes'].sum()
plt.figure(figsize=(5,5))
plt.title('City data votes who price is more than 3.0')
y.plot.pie(autopct='%1.2f%%')
plt.show()
```

```python
#Compare Agra, Ahmedabad and gandhinagr how much restaurant have less than 1.0 Price and RATING is 4.0.
x=df[((df['city']=='Agra') | (df['city']=='Ahmedabad') | (df['city']=='Gandhinagar')) | (df['price_range']<1.0) & (df['aggregate_rating']>4.0)]
x['city'].value_counts().plot.bar()
plt.xlabel("City")
plt.ylabel("Number of Restaurant")
plt.show()
```

```python
# In Tajganj, Thaltej and Civil Lines Region which restaurant get maximum Votes.
x=df[(df['locality']=='Tajganj') | (df['locality']=='Thaltej') | (df['locality']=='Civil Lines') | (df['locality']=='Ambavadi') | (df['locality']=='Prahlad nagar')]
y=x.groupby(['locality'])['votes'].sum()
plt.figure(figsize=(5,5))
plt.title('REGION data votes who price is more than 3.0')
y.plot.pie(autopct='%1.2f%%')
plt.show()
```
